```python
#PLOTTING HARE-LYNX-CARROT DATA USING MATPLOTLIB
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('populations.txt')
year,hare,lynx,carrot = data.T

print('YEAR    HARE    LYNX    CARROT\n', data)
print('\n***********************************************************\n')
print('YEAR\n',year)
print('\n***********************************************************\n')
print('HARE\n',hare)
print('\n***********************************************************\n')
print('LYNX \n', lynx)
print('\n***********************************************************\n')
print('CARROT\n', carrot)
print('\n***********************************************************\n')

plt.figure(1, figsize = (10,5))
plt.plot(year,hare,color='green', label='^HARE')
plt.plot(year,lynx,color='red', label='^LYNX')
plt.plot(year,carrot,color='orange', label='^CARROT')
plt.legend()
plt.grid()


#SUB PLOTS

plt.figure(2, figsize=(12,7))

plt.subplot(221)
plt.plot(year,hare,color='green', label='^HARE')
plt.plot(year,lynx,color='red', label='^LYNX')
plt.plot(year,carrot,color='orange', label='^CARROT')
plt.legend()
plt.grid()

plt.subplot(222)
plt.plot(year,hare,color='green', label='^HARE')
plt.legend()
plt.grid()


plt.subplot(223)
plt.plot(year,lynx,color='red', label='^LYNX')
plt.legend()
plt.grid()

plt.subplot(224)
plt.plot(year,carrot,color='orange', label='^CARROT')
plt.legend()
plt.grid()

plt.show()
```
